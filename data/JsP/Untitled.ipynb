```python
import pandas as pd
import xlrd
from googletrans import Translator
import time
translator = Translator()

df =pd.read_excel('논문검색리스트 정제 (1).xls')



# 이걸 키워드 찢기 전에 먼저 해야댐 
title_ko = df['논문명']
title_en = df['논문 외국어명']

# translator.detect(title_ko[282]).lang
for page in range(0,6):
    for i in range(page*50,(page+1)*50):
        tmp=translator.detect(title_ko[i]).lang
        print(i,tmp)
        if translator.detect(df['저자'][i]).lang == 'ko':
            df.loc[i,'저자'] = translator.translate(df['저자'][i],dest='en').text
        if pd.isna(df['공동저자'][i])==False and translator.detect(df['공동저자'][i]).lang == 'ko':
            df.loc[i,'공동저자'] = translator.translate(df['공동저자'][i],dest='en').text
        if pd.isna(df['키워드(한국어)'][i])==False and translator.detect(df['키워드(한국어)'][i]).lang == 'ko':
            df.loc[i,'키워드(한국어)'] = translator.translate(df['키워드(한국어)'][i],dest='en').text
        
        if tmp=='en': # 한글 논문명이 영어일때
            if pd.isna(title_en[i]):# 외국어명 비어있으면 
                df.loc[i,'논문 외국어명']=title_ko[i]
                df.loc[i,'논문명']=translator.translate(title_ko[i],dest='ko').text
    #             print(i,df.loc[i,'논문명'])
            else: # 안비어있고 
                if translator.detect(title_en[i]).lang =='en':#영어면
                    df.loc[i,'논문명']=translator.translate(title_ko[i],dest='ko').text
    #                 print(i,'#')
    #             else:#g한국어면
        else: # 한글 논문명이 한글일때 
            if pd.isna(title_en[i]):# 외국어명 비어있으면 
                df.loc[i,'논문 외국어명']=translator.translate(title_ko[i],dest='en').text
    #             print(i,'%')
    
    
    
    
# #     time.sleep(0.5)
# #     print(tmp)
#     if tmp == 'en':
#         df.loc[i,'논문 외국어명']=title_ko[i]
#         df.loc[i,'논문명']= translator.translate(title_ko[i],dest='ko').text
#         time.sleep(0.5)
#         print(i,translator.translate(title_ko[i],dest='ko').text)
#     else:
#         df.loc[i,'논문 외국어명'] = translator.translate(title_ko[i],dest='en').text
#         time.sleep(0.5)
#         print(i,translator.translate(title_ko[i],dest='en').text)
df.tail(20)
```
